## Skysat Bundle Adjustment Pipeline 

Allows to adjust a set of views and add new views to a set of already adjusted images.

### Experiment name 

```python
exp_name = 'skysat_L1A/dona_juana2_9'
input_dir = os.path.join('exp/',exp_name)
```

### 1. Load previously adjusted data (if existent)

```python
n_adj = 0
myimages_adj, mycrops_adj, P_crop_adj, myrpcs_adj = [], [], [], []
if os.path.isdir(input_dir + '/frames_adj'):

    # read tiff images 
    myimages_prev = pickle.load(open(input_dir+'/myimages.pickle','rb')) 
    myimages_adj = [input_dir + '/frames_adj/' + image_fn for image_fn in myimages_prev]
    n_adj += len(myimages_adj)
    print('Found {} previously adjusted images'.format(n_adj))

    # get image crops
    mycrops_adj = []
    for image in myimages_adj:
        tmp = utils.readGTIFF(image)[:,:,0]
        mycrops_adj.append({ 'crop': utils.simple_equalization_8bit(tmp), 'x0': 0.0, 'y0': 0.0 })

    # load previously adjusted projection matrices
    cam_model = 'Affine'
    P_crop_adj = []
    for it_cont in np.arange(n_adj):   
        with open(os.path.splitext(myimages_adj[it_cont])[0].replace('frames_adj','P_adj')+'_pinhole_adj.json','r') as f:
            P_img = np.array(json.load(f)['P_projective'])
        P_crop_adj.append(P_img/P_img[2,3])
        print('\rCamera {} / {} done'.format(it_cont+1, n_adj), end = '\r')
    print('\nProjection matrices read')

    # load previously adjusted rpc
    myrpcs_adj = [rpcm.rpc_from_rpc_file(os.path.splitext(im)[0].replace('frames_adj', 'RPC_adj')+'_RPC_adj.txt') for im in myimages_adj]
    print('Finished loading RPC coefficients')
    
    # load bundle adjusted 3d point locations

else:
    print('No previously adjusted views were found!')
```

### 2. Load new view(s)

```python
myimages_new = np.sort(glob.glob(input_dir+'/frames/*.tif')).tolist()
n_new = len(myimages_new)
print('Found {} new images'.format(n_new))

# load new images
mycrops_new = []
for image in myimages_new:
    tmp = utils.readGTIFF(image)[:,:,0]
    mycrops_new.append({ 'crop': utils.simple_equalization_8bit(tmp), 'x0': 0.0, 'y0': 0.0 })
    
# load associated projection matrices
cam_model = 'Perspective'
P_crop_new = []
for iter_cont in np.arange(n_new):
    with open(os.path.splitext(myimages_new[iter_cont])[0]+'_pinhole.json', 'r') as f:
        P_img = np.array(json.load(f)['P_projective'])
    P_crop_new.append(P_img/P_img[2,3])
    print('\rCamera {} / {} done'.format(iter_cont+1, n_new), end = '\r')
print('\nProjection matrices read')

# load associated rpcs
myrpcs_new = [rpcm.rpc_from_rpc_file(os.path.splitext(im)[0]+'_RPC.TXT') for im in myimages_new]
print('Finished loading RPC coefficients')

# update myimages list
myimages = myimages_adj + myimages_new
pickle_out = open(input_dir+'/myimages.pickle','wb')
pickle.dump([os.path.basename(imagefn) for imagefn in myimages], pickle_out)
pickle_out.close()
```

### 3. Match new view(s) to previous images

```python
input_crops = mycrops_adj + mycrops_new
input_seq = [input_crops[f]['crop'] for f in range(len(input_crops))]    
input_P = P_crop_adj + P_crop_new
input_rpcs = myrpcs_adj + myrpcs_new
n_img = len(input_seq)
footprints = ba_utils.get_image_footprints(input_rpcs, input_seq)
```

### 4. Feature detection

```python
# FEATURE DETECTION + MATCHING ON THE NEW IMAGES
import timeit

print('Running feature detection...\n')
start = timeit.default_timer()
last_stop = start

kp_cont, features = 0, [] 

# load previous feature tracks if existent
if n_adj > 0:
    pickle_in = open(input_dir+'/features.pickle','rb')
    features = pickle.load(pickle_in)
    kp_cont = features[-1]['id'][-1] + 1
    print('Previous features loaded!')

# feature detection on the new view(s)
new_indices = np.arange(n_adj, n_adj + n_new)
new_input_seq = [input_seq[idx] for idx in new_indices]
new_input_rpcs = [input_rpcs[idx] for idx in new_indices]
new_footprints = [footprints[idx] for idx in new_indices]
new_features = ba_utils.feature_detection_skysat(new_input_seq, new_input_rpcs, new_footprints)
for current_features in new_features:
    current_features['id'] += kp_cont
features.extend(new_features)

pickle_out = open(input_dir+'/features.pickle','wb')
pickle.dump(features, pickle_out)
pickle_out.close()
print('\nDetected features saved!')
stop = timeit.default_timer()
print('\n...done in {} seconds'.format(stop - last_stop))
last_stop = stop

print('\nComputing pairs to be matched...\n')

all_pairwise_matches, pairs_to_match, pairs_to_triangulate = [], [], []

# load previous matches and list of paris to be matched/triangulate if existent
if n_adj > 0:
    pickle_in = open(input_dir+'/matches.pickle','rb')
    all_pairwise_matches, pairs_to_match, pairs_to_triangulate = pickle.load(pickle_in)
    print('Previous matches loaded!')

# possible new pairs to match are composed by 1 + 2 
# 1. each of the previously adjusted images with the new ones
possible_pairs = []
for i in np.arange(n_adj):
    for j in np.arange(n_adj, n_adj + n_new):
        possible_pairs.append((i, j))       
# 2. each of the new images with the rest of the new images
for i in np.arange(n_adj, n_adj + n_new):
    for j in np.arange(i+1, n_adj + n_new):
        possible_pairs.append((i, j))
        
# filter stereo pairs that are not overlaped
# stereo pairs with small baseline should not be used to triangulate 
pairs2match, pairs2triangulate = ba_utils.filter_pairs_to_match_skysat(possible_pairs, footprints, input_P)
pairs_to_match.extend(pairs2match)
pairs_to_triangulate.extend(pairs2triangulate)
print('{} new pairs to be matched'.format(len(pairs2match)))

stop = timeit.default_timer()
print('\n...done in {} seconds'.format(stop - last_stop))
last_stop = stop

print('\nMatching...\n')
new_pairwise_matches = ba_utils.matching_skysat(pairs2match, features) 
all_pairwise_matches.extend(new_pairwise_matches)
        
pickle_out = open(input_dir+'/matches.pickle','wb')
pickle.dump([all_pairwise_matches, pairs_to_match, pairs_to_triangulate], pickle_out)
pickle_out.close()
print('\nPairwise matches saved!')      
stop = timeit.default_timer()
print('\n...done in {} seconds'.format(stop - last_stop))
last_stop = stop

print('\nBuilding feature tracks...\n')
C = ba_utils.feature_tracks_from_pairwise_matches(features, all_pairwise_matches, pairs_to_triangulate)
pickle_out = open(input_dir+'/Cmatrix.pickle','wb')
pickle.dump(C, pickle_out)
pickle_out.close()
print('\nCorrespondence matrix saved!')
stop = timeit.default_timer()
print('\n...done in {} seconds'.format(stop - last_stop))
last_stop = stop

hours, rem = divmod(last_stop - start, 3600)
minutes, seconds = divmod(rem, 60)
print('\nTotal time: {:0>2}:{:0>2}:{:05.2f}'.format(int(hours),int(minutes),seconds))
```

### Define Bundle Adjustment parameters

```python
'''
INPUT PARAMETERS FOR BUNDLE ADJUSTMENT
'cam_params': (n_cam, 12), initial projection matrices. 1 row = 1 camera estimate.
              first 3 elements of each row = R vector, next 3 = T vector, then f and two dist. coef.
'pts_3d'    : (n_pts, 3) contains the initial estimates of the 3D points in the world frame.
'cam_ind'   : (n_observations,), indices of cameras (from 0 to n_cam - 1) involved in each observation.
'pts_ind'   : (n_observations,) indices of points (from 0 to n_points - 1) involved in each observation.
'pts_2d'    : (n_observations, 2) 2-D coordinates of points projected on images in each observations.
'''

print('Defining BA input parameters...')
params_opt, cam_params, pts_3d, pts_2d, cam_ind, pts_ind, ba_params \
= ba_utils.set_ba_params(input_P, C, cam_model, n_adj, n_new, pairs_to_triangulate)
print('...done!\n')

print('pts_2d.shape:{}  pts_ind.shape:{}  cam_ind.shape:{}'.format(pts_2d.shape, pts_ind.shape, cam_ind.shape))
print('pts_3d.shape:{}  cam_params.shape:{}\n'.format(pts_3d.shape, cam_params.shape))
print('Bundle Adjustment parameters defined')

if ba_params['n_params'] > 0 and ba_params['opt_X']:
    print('  -> Both camera parameters and 3D points will be optimized')
elif ba_params['n_params'] > 0 and not ba_params['opt_X']:
    print('  -> Only the camera parameters will be optimized')
else:
    print('  -> Only 3D points will be optimized')
```

### Run Bundle Adjustment (soft L1 loss)

```python
import time
from scipy.optimize import least_squares, minimize

pts_2d_w = np.ones(pts_2d.shape[0])

# compute loss value and plot residuals at the initial parameters

f0 = ba_utils.fun(params_opt, cam_ind, pts_ind, pts_2d, cam_params, pts_3d, ba_params, pts_2d_w)
plt.plot(f0)

# define jacobian
A = ba_utils.bundle_adjustment_sparsity(cam_ind, pts_ind, ba_params)

# run bundle adjustment
t0 = time.time()
res = least_squares(ba_utils.fun, params_opt, jac_sparsity=A, verbose=1, x_scale='jac',
                    method='trf', ftol=1e-4, xtol=1e-10, loss='soft_l1', f_scale = 0.5,
                    args=(cam_ind, pts_ind, pts_2d, cam_params, pts_3d, ba_params, pts_2d_w))

t1 = time.time()
print("Optimization took {0:.0f} seconds\n".format(t1 - t0))

#plot residuals at the found solution
plt.plot(res.fun);

# recover BA output
pts_3d_ba, cam_params_ba, P_crop_ba = ba_utils.get_ba_output(res.x, ba_params, cam_params, pts_3d)

# check BA error performance
ba_e = ba_utils.check_ba_error(f0, res.fun, pts_2d_w)
```

### Remove outliers according to reprojection error

```python
elbow_value = ba_utils.get_elbow_value(ba_e, 95)
fig = plt.figure()
plt.plot(np.sort(ba_e))
plt.axhline(y=elbow_value, color='r', linestyle='-')
plt.show()
newC = ba_utils.remove_outlier_obs(ba_e, pts_ind, cam_ind, C, pairs_to_triangulate, thr=max(elbow_value,2.0))

pickle_out = open(input_dir+'/Cmatrix2.pickle','wb')
pickle.dump(newC, pickle_out)
pickle_out.close()
print('Correspondence matrix saved!')

print('Defining BA input parameters...')
params_opt, cam_params, pts_3d, pts_2d, cam_ind, pts_ind, ba_params \
= ba_utils.set_ba_params(input_P, newC, cam_model, n_adj, n_new, pairs_to_triangulate)
print('...done!\n')
```
